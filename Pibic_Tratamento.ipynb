Importing de CSV file

```python

# What I'm doing here: importing the CSV file and loading it into a pandas DataFrame

df = pd.read_csv("coleta.csv.csv")
df.head()
```

To work with yfinance, we need to add the ".sa" at the end of each stock

```python


df["Cod"] = df["Cod"] + ".SA"
df.head()
```

Getting the data from the range (2020 - 2024), also listing the stock that we dont have data

```python

start_date = '2000-01-01'
end_date = '2024-12-31'

ohlcv_data = {}

empty_tickets = []
map_setor = {}


for ticker in df['Cod']:

    try:
            
        data = yf.download(tickers=ticker, start=start_date, end=end_date)
        ticker_obj = yf.Ticker(ticker)
        info = ticker_obj.info
        sector = info.get('sector', 'N/A') or info.get('industry') or 'N/A'
        map_setor[ticker] = sector



        if data.empty:
            empty_tickets.append(ticker)
            print(f"Ticket : {ticker} no Data")

        ohlcv_data[ticker] = data

    except Exception as e:

        empty_tickets.append(ticker)
        print(f"{e}")
```

Data Handling

```python


for ticker, data in ohlcv_data.items():
    if isinstance(data.columns, pd.MultiIndex):
        try:
            data.columns = data.columns.droplevel(level=1)
        except IndexError:
            pass
            
    ohlcv_data[ticker] = data
```

```python
print(ohlcv_data["PETR4.SA"])
```

We are regrouping the data so we can save the information for each quarter of the stock during the 2020–2024 period

```python
def aggregate_to_3(data):

    data.index = pd.to_datetime(data.index)

    aggregation_rules = {
        'Open': 'first',      
        'High': 'max',        
        'Low': 'min',          
        'Close': 'last',       
        'Volume': 'sum'       
    }

    quarterly_data = data.resample("QE").agg(aggregation_rules)

    return quarterly_data

df_quart = {}

for ticker, data_df in ohlcv_data.items():

    df_quart[ticker] = aggregate_to_3(data_df)

print(df_quart["PETR4.SA"])
```

Previously, we had the data stored in a dictionary of DataFrames. This cell handles the data conversion to a single DataFrame.

```python
for ticker, df in df_quart.items():
    df["Ticker"] = ticker

final_quarterly_df = pd.concat(df_quart.values())



final_quarterly_df['Sector'] = final_quarterly_df["Ticker"].map(map_setor)
```

```python
final_quarterly_df.to_csv('quarterly_financial_data.csv')
```

 Now, for each feature in OHCLV, we are going to get the following statistical measures: mean, median, standard deviation, maximum, minimum. It is organized in the following way: Dictionary[ticker][statistical measure] = []
 It means, there is a list for each feature 

```python
stats_df = (
    final_quarterly_df.groupby('Ticker')[['Open', 'High', 'Low', 'Close', 'Volume']]
      .agg(['mean', 'median', 'max', 'min', 'std'])
      .reset_index()
)
```

Now we have a DataFrame containing all the information from all the statistical measures, from all the columns

```python
final_quarterly_df = pd.read_csv('quarterly_financial_data.csv')
final_quarterly_df["Return"] = 100*((final_quarterly_df['Close'] - final_quarterly_df['Open']) / final_quarterly_df['Open']) 

correlations = (
    final_quarterly_df.groupby(['Ticker', 'Sector'])
      .apply(lambda x: x['Return'].corr(x['Volume']))
      .reset_index(name='Correlation')
)



top5 = correlations.reindex(correlations['Correlation'].abs().sort_values(ascending=False).index).head(5)
top5_positive = correlations.sort_values('Correlation', ascending=False).head(5)
top5_negative = correlations.sort_values('Correlation', ascending=True).head(5)
print(top5_positive)
print(top5_negative)
```

Graph visutalization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

top5 = (
    final_quarterly_df.groupby('Ticker')['Return']
    .mean()
    .nlargest(7)
    .index
)
top5=top5[2:]

# Filter data for only those tickers

top5_df = final_quarterly_df[final_quarterly_df['Ticker'].isin(top5)]

# Plot time series
plt.figure(figsize=(10,6))
sns.lineplot(data=top5_df, x='Date', y='Return', hue='Ticker', marker='o')
plt.title('Quarterly Returns — Top 5 Assets with Highest Average Return')
plt.xlabel('Quarter')
plt.ylabel('Return (%)')
plt.legend(title='Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

bottom5 = (
    final_quarterly_df.groupby('Ticker')['Return']
    .mean()
    .nsmallest(5)
    .index
)


top5_df = final_quarterly_df[final_quarterly_df['Ticker'].isin(bottom5)]


plt.figure(figsize=(10,6))
sns.lineplot(data=top5_df, x='Date', y='Return', hue='Ticker', marker='o')
plt.title('Quarterly Returns — Top 5 Assets with Lowest Average Return')
plt.xlabel('Quarter')
plt.ylabel('Return (%)')
plt.legend(title='Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_setores = final_quarterly_df['Sector'].value_counts().reset_index()
df_setores.columns = ['Sector', 'Count']

plt.figure(figsize=(10,6))
sns.barplot(data=df_setores, x='Sector', y='Count', palette='Set2')
plt.title('Number of Records by Sector')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
